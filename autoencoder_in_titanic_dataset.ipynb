```python
x_train,x_test,y_train,y_test=train_test_split(dt3.drop('Survived',axis=1),dt3.iloc[:,0],test_size=0.1)      
type(y_train)           
















```

```python
x_train  

```

```python
scaler=StandardScaler()

x_test = np.array(x_test) 
x_train = np.array(x_train)
x_train[:, [0, 1, 2]] = scaler.fit_transform(x_train[:, [0, 1, 2]])  
x_test[:, [0, 1, 2]] = scaler.transform(x_test[:, [0, 1, 2]])             
x_train.shape[1]  














```

```python
def custom_clustering_loss(y_true, y_pred):
    mse_loss = tf.reduce_mean(tf.square(y_true - y_pred[0]))
    q = y_pred[1]
    p = tf.reduce_sum(q, axis=0) / tf.reduce_sum(q)
    kl_loss = tf.reduce_sum(p * tf.math.log(p / q))
    
    return mse_loss + kl_loss   
import tensorflow as tf 
from tensorflow import keras 
from keras.models import Model
from keras.layers import Input, Dense
input_dim = x_train.shape[1]

input_layer = Input(shape=(input_dim,))
encoder = Dense(1024, activation='relu')(input_layer) 
layer_2=Dense(512,activation='relu')(encoder)  
layer_3=Dense(256,activation='relu')(layer_2) 
layer_4=Dense(128,activation='relu')(layer_3) 
layer_5=Dense(64,activation='relu')(layer_4) 
layer_6=Dense(32,activation='relu')(layer_5) 
layer_7=Dense(16,activation='relu')(layer_6)  
layer_8=Dense(8,activation='relu')(layer_7)



decoder = Dense(input_dim, activation='sigmoid')(layer_8)

autoencoder = Model(inputs=input_layer, outputs=decoder)
autoencoder.compile(optimizer='adam', loss=custom_clustering_loss)           











```

```python
autoencoder.fit(x_train, x_train, epochs=100, shuffle=True, validation_data=(x_test, x_test))            











```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=5, random_state=42)  
encoded_data=autoencoder.predict(x_train) 
encoded_data_test=autoencoder.predict(x_test)

clusters = kmeans.fit_predict(encoded_data) 
clusters_test=kmeans.predict(encoded_data_test)
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score 
clf = RandomForestClassifier(n_estimators=50, random_state=42)  



clf.fit(clusters.reshape(-1,1),y_train)    
y_pred=clf.predict(clusters_test.reshape(-1,1))





accuracy = accuracy_score(y_pred,y_test) 
accuracy                   




















```

```python
from sklearn.model_selection import cross_val_score
cross_val_score(clf, y_test.to_numpy().reshape(-1,1),y_pred.reshape(-1,1), cv=10, scoring='accuracy').mean()         















```
